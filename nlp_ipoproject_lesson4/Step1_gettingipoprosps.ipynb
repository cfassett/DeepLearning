```python
def getedgarsfromticker(symbol, ciks, edgaridx):
    ## This function takes a symbol and returns the dataframe in the edgar index file for the matching CIK
    edgarcik=ciks[ciks['Ticker']==symbol]["CIK"].values[0]  ##Pull the cik value for a given symbol
    
    return edgaridx[edgaridx['code']==edgarcik]
```

```python
nufodf=getedgarsfromticker("NUFO",ciks,edgaridx)
```

```python
nufodf[nufodf['form']=="424B4"].iloc[0]['html']
```

```python
def getedgarprospectusfromedgarsubdf(pandasubdf):
    # edgar url + first prospectus in the subdf
    html="https://www.sec.gov/Archives/"+pandasubdf[pandasubdf['form']=="424B4"].iloc[0]['file']
    
    return html
```
